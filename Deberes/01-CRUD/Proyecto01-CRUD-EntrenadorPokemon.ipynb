```python
# Menu Principal
def menu_prinicipal():  
    print("1. Entrenador")
    print("2. Pokemon")
    opcion_entidad = input("Escoga la entidad a trabajar: ")
    
    return opcion_entidad

# Menu CRUD
def menu_generico():
    print("1. Consultar")
    print("2. Ingresar")
    print("3. Actualizar")
    print("4. Eliminar")
    opcion_seleccionar_crud = input("Escoga la opcion: ")
   
    return opcion_seleccionar_crud

entidad = menu_prinicipal()
operacion_crud = menu_generico()

print(entidad,operacion_crud)
```

```python
# CRUD - buscar
def crud_buscar(id_a_buscar, entidad):
    if (entidad == "1"):
        try:
            path = "./Entrenador.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
                   
            for linea in contenido:
                id_encontrado = linea.find(id_a_buscar)
                
                if (id_encontrado == 0):
                    print(linea)
            
            archivo_abierto.close()
        except Exception as error:
            print("Error")
    
    else:
        try:
            path = "./Pokemon.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
            
            for linea in contenido:
                id_encontrado = linea.find(id_a_buscar)
            
                if (id_encontrado == 0):
                    print(linea)
            
            archivo_abierto.close()
        except Exception as error:
            print("Error")
    
    return archivo_abierto


# CRUD - insertar
def crud_insertar(datos, entidad):
    if (entidad == "1"):
        try:
            path = "./Entrenador.txt"
            archivo_abierto = open(path, mode = "a")
                               
            archivo_abierto.write(f"\n{datos}")
            
            archivo_abierto.close()
        except Exception as error:
            print("Error")
    
    else:
        try:
            path = "./Pokemon.txt"
            archivo_abierto = open(path, mode = "a")
                        
            archivo_abierto.write(f"\n{datos}")
            
            archivo_abierto.close()
        except Exception as error:
            print("Error")
    
    return archivo_abierto

# CRUD - eliminar
def crud_eliminar(id_a_eliminar, entidad):
    if (entidad == "1"):
        try:
            path = "./Entrenador.txt"
            archivo_abierto = open(path, "r")
            contenido = archivo_abierto.readlines()
            
            archivo_abierto = open(path, "w")
                            
            for linea in contenido:
                id_encontrado = linea.find(id_a_eliminar)
                if (id_encontrado != 0):
                    archivo_abierto.write(linea)
                            
            archivo_abierto.close()
        except Exception as error:
            print("Error")
    
    else:
        try:
            path = "./Pokemon.txt"
            archivo_abierto = open(path, "r")
            contenido = archivo_abierto.readlines()
            
            archivo_abierto = open(path, "w")
                            
            for linea in contenido:
                id_encontrado = linea.find(id_a_eliminar)
                if (id_encontrado != 0):
                    archivo_abierto.write(linea)
            
            archivo_abierto.close()
        except Exception as error:
            print("Error")
    
    return archivo_abierto

# Imprimir Archivo
def imprimir_archivo(entidad):
    if(entidad == "1"):
        try:
            path = "./Entrenador.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
        
            for linea in contenido:
                print(linea)
        
            archivo_abierto.close()
        except Exception as error:
            print("Error")
    
    else:
        try:
            path = "./Pokemon.txt"
            archivo_abierto = open(path)
            contenido = archivo_abierto.readlines()
        
            for linea in contenido:
                print(linea)
        
            archivo_abierto.close()
        except Exception as error:
            print("Error")

            
# CRUD - actualizar
def crud_actualizar(id_a_actualizar, entidad):
    if (entidad == "1"):
        try:
            linea_actualizada = input("Ingrese la linea actualizada: ")
            
            path = "./Entrenador.txt"
            archivo_abierto = open(path, "r")
            contenido = archivo_abierto.readlines()
            archivo_abierto.close()
            
            archivo_abierto = open(path, "w")
            
            print(contenido)
                        
            for linea in contenido:
                id_encontrado = linea.find(id_a_actualizar)
                if (id_encontrado != 0):
                    archivo_abierto.write(linea)
                else:
                    archivo_abierto.write(linea_actualizada)
                                         
            archivo_abierto.close()
        except Exception as error:
            print("Error")
    
    else:
        try:
            linea_actualizada = input("Ingrese la linea actualizada: ")
            
            path = "./Pokemon.txt"
            archivo_abierto = open(path, "r")
            contenido = archivo_abierto.readlines()
            archivo_abierto.close()
            
            archivo_abierto = open(path, "w")
            
            print(contenido)
                        
            for linea in contenido:
                id_encontrado = linea.find(id_a_actualizar)
                if (id_encontrado != 0):
                    archivo_abierto.write(linea)
                else:
                    archivo_abierto.write(linea_actualizada)
                                         
            archivo_abierto.close()
        except Exception as error:
            print("Error")
    
    return archivo_abierto            

# CRUD Opciones
if (operacion_crud == "1"):
    id_a_buscar = input("Introduzca el id a buscar: ")
    crud_buscar(id_a_buscar, entidad)

if(operacion_crud == "2"):
    if(entidad == "1"):
        print("Ingrese los siguientes campos: ")
        datos = input("Id Nombre Apellido Numero-Medallas Campeon? ")
        crud_insertar(datos, entidad)
    
    else:
        print("Ingrese los siguientes campos: ")
        datos = input("Id Nombre Tipo Poder-Especial Nivel idEntrenador ")
        crud_insertar(datos, entidad)

if (operacion_crud == "3"):
    imprimir_archivo(entidad)
    id_a_actualizar = input("Introduzca el id a actualizar: ")
    crud_actualizar(id_a_actualizar, entidad)
        
if (operacion_crud == "4"):
    id_a_eliminar = input("Introduzca el id a eliminar: ")
    crud_eliminar(id_a_eliminar, entidad)
    
```
