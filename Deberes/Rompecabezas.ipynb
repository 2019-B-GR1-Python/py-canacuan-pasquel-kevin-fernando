```python
import os
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
```

```python
mapache = misc.face()
mapache_parts = np.vsplit(mapache,4)
piezas_index = np.random.permutation(4)
mapache_desordenado = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))
plt.imshow(mapache_desordenado)
plt.show()

def menu():
    os.system('cls') 
    print ("Selecciona una opción:")
    print ("\t1 - Armar rompecabezas...")
    print ("\t2 - Reiniciar rompecabezas...")
    print ("\t3 - Salir...")

while True:
    menu()
    opcionMenu = input("Ingrese un valor: ")
    
    if opcionMenu=="1":
        mover_posicion_a = input("Seleccione una pieza(0-3): ")
        mover_posicion_b = input("Selecciona la posicion con la que desea intercambiar la pieza seleccionada ")
        pieza_temporal = piezas_index[int(mover_posicion_a)]
        piezas_index[int(mover_posicion_a)] = piezas_index[int(mover_posicion_b)]
        piezas_index[int(mover_posicion_b)] = pieza_temporal
        mapache_desordenado = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))
        plt.imshow(mapache_desordenado)
        plt.show()
    
    elif opcionMenu=="2":
        piezas_index = np.random.permutation(4)
        mapache_desordenado = np.concatenate(list(map(lambda x: mapache_parts[x],piezas_index)))
        plt.imshow(mapache_desordenado)
        plt.show()
        
    elif opcionMenu=="3":
        break
    else:
        print ("")
        input("No has pulsado ninguna opción correcta...\npulsa una tecla para continuar")
```
