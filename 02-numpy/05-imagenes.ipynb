```python
import numpy as np
from scipy import ndimage
from scipy import misc
```

```python
mapache = misc.face()
mapache
```

```python
mapache.shape
```

```python
type(mapache)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mapache)
plt.show()
```

```python
plt.imshow(mapache[::-1])
```

```python
blurred_mapache = ndimage.gaussian_filter(mapache, sigma=5)

plt.imshow(blurred_mapache)
```

```python
mapache_cambiado = mapache.reshape(1024,768,3)

plt.imshow(mapache_cambiado)
plt.show()
```

```python
mapache_mitad1, mapache_mitad2 = np.split(mapache,2)

plt.imshow(mapache_mitad1)
plt.show()

plt.imshow(mapache_mitad2)
plt.show()
```

```python
mapache_rojo = mapache[:,:,1] + 100

plt.imshow(mapache_rojo)
plt.show()
```

```python
mapache_cuadro = mapache[:600,:600]

plt.imshow(mapache_cuadro)
plt.show()
```
